벡터연산

```python
import numpy as np

# 벡터 생성
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

# 벡터 덧셈 및 뺄셈
add = v1 + v2
subtract = v1 - v2

# 벡터 내적 (dot product)
dot_product = np.dot(v1, v2)

# 벡터 크기 (magnitude)
magnitude_v1 = np.linalg.norm(v1)

print("Vector Addition:", add)
print("Vector Subtraction:", subtract)
print("Dot Product:", dot_product)
print("Magnitude of v1:", magnitude_v1)
```

행열연산


```python
# 행렬 생성
matrix_a = np.array([[1, 2], [3, 4]])
matrix_b = np.array([[2, 0], [1, 3]])

# 행렬 덧셈, 뺄셈, 스칼라 곱
add = matrix_a + matrix_b
subtract = matrix_a - matrix_b
scalar_multiply = 3 * matrix_a

# 행렬 내적
matrix_multiply = np.dot(matrix_a, matrix_b)

# 원소별 곱셈
elementwise_multiply = matrix_a * matrix_b

print("Matrix Addition:\n", add)
print("Matrix Subtraction:\n", subtract)
print("Scalar Multiplication:\n", scalar_multiply)
print("Matrix Multiplication:\n", matrix_multiply)
print("Element-wise Multiplication:\n", elementwise_multiply)
```

행렬의 성질

```python
matrix_a
```

```python
# 행렬의 전치
transpose = np.transpose(matrix_a)
print(transpose)

# 행렬의 역행렬
try:
    inverse = np.linalg.inv(matrix_a) # 역행렬: 주어진 행렬을 곱했을 때 단위 행렬이 되는 행렬을 구합니다.
    print('matrix_a :',matrix_a)
    print("Inverse of Matrix A:\n", inverse)
except np.linalg.LinAlgError:
    print("Matrix A is not invertible.")

# 행렬식 (Determinant)
determinant = np.linalg.det(matrix_a)
print("Determinant of Matrix A:", determinant)

# Trace (대각선 성분의 합)
trace = np.trace(matrix_a)
print("Trace of Matrix A:", trace)
```

```python
import numpy as np

# 행렬 A 정의
A = np.array([[4, 7], [2, 6]])

# 역행렬 계산
try:
    A_inv = np.linalg.inv(A)  # np.linalg.inv: 역행렬 계산
    print("Matrix A:")
    print(A)
    print("\nInverse of A:")
    print(A_inv)
except np.linalg.LinAlgError:
    print("Matrix A is singular and does not have an inverse.")
```

Norm과 거리 계산

```python
# 벡터 Norm
# Norm: 벡터나 행렬의 크기를 측정합니다.
v = np.array([3, 4])
norm_2 = np.linalg.norm(v)  # L2 Norm  # L2 Norm (유클리드 거리
norm_1 = np.linalg.norm(v, 1)  # L1 Norm  L1 Norm (맨해튼 거리)

# 두 점 사이의 유클리드 거리
point_a = np.array([1, 2])
point_b = np.array([4, 6])
euclidean_distance = np.linalg.norm(point_a - point_b)

print("L2 Norm of v:", norm_2)
print("L1 Norm of v:", norm_1)
print("Euclidean Distance between points:", euclidean_distance)

# Norm은 벡터의 크기 또는 행렬의 성질을 나타냅니다. 거리는 데이터 포인트 간의 유사성 측정에 유용합니다.
```

프로젝션과 직교화

```python
import numpy as np

# 벡터 v1을 v2에 투영
v1 = np.array([3, 4])
v2 = np.array([1, 2])
projection = (np.dot(v1, v2) / np.dot(v2, v2)) * v2

# 직교화 (Gram-Schmidt Process)
def gram_schmidt(vectors):
    orthogonal = []
    for v in vectors.astype(float):  # 실수형 변환
        for u in orthogonal:
            v -= np.dot(v, u) / np.dot(u, u) * u
        orthogonal.append(v / np.linalg.norm(v))
    return np.array(orthogonal)

# 벡터 집합
vectors = np.array([[1, 1], [1, -1]], dtype=float)  # 실수형으로 선언
orthogonal_basis = gram_schmidt(vectors)

print("Projection of v1 onto v2:", projection)
print("Orthogonal Basis:\n", orthogonal_basis)
```
