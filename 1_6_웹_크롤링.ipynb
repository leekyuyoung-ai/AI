# 할리스 커피 매장정보 수집하기

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

df_lists = []
for i in range(1,50):
  url = f'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={i}'
  response = requests.get(url)
  response.raise_for_status()  # 요청이 성공했는지 확인
  # HTML 파싱
  soup = BeautifulSoup(response.text, 'html.parser')  # 웹사이트의 정보를 가져온다.. sample.html 파일 내용처럼
  # datas는 첫페이지의 table을 구성하는 row 들의 집합 -- tr의 집합
  datas = soup.select('#contents > div.content > fieldset > fieldset > div.tableType01 > table > tbody > tr')

  result = []
  for i in range( len(datas) ):  # tr은 td들의 집합
    region = datas[i].select('td')[0].text
    storename = datas[i].select('td')[1].text
    state = datas[i].select('td')[2].text
    addr = datas[i].select('td')[3].text
    tel = datas[i].select('td')[5].text
    result.append([region, storename, state, addr, tel])

  df_lists.append(pd.DataFrame(result))

# 모든매장의 정보를 df로 저장한 리스트들을  합친다
stores = pd.concat(df_lists).reset_index(drop=True)  # 인덱스 정렬
stores.columns = ['지역','매장명','매장타입','주소','전화번호']   # 컬럼 제목을 넣어준다
stores.to_csv('hollys_stores.csv',encoding='UTF-8') # 파일로 저장한다
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re

result = []
for i in range(1,11):
  url = f'https://www.police.go.kr/user/bbs/BD_selectBbsList.do?q_bbsCode=1013&q_bbscttSn=&q_tab=&q_searchKeyTy=&q_searchVal=&q_rowPerPage=10&q_currPage={i}'
  response = requests.get(url)
  response.raise_for_status()  # 요청이 성공했는지 확인
  # HTML 파싱
  soup = BeautifulSoup(response.text, 'html.parser')  # 웹사이트의 정보를 가져온다.. sample.html 파일 내용처럼
  # datas는 첫페이지의 table을 구성하는 row 들의 집합 -- tr의 집합
  datas = soup.select('#contents2 > div > div.col-md-9.contents > div.block.table-responsive.grid-list.for-mobile > table > tbody > tr')
  title = datas[0].select('td.subject')[0].text # 제목만 추출
  title = re.sub('[^가-힣]+',' ',title).strip() # 한글을 제외한 나머지는 공백으로, 문장의 앞뒤의 공백을 제거

  link_url = 'https://www.police.go.kr/user/bbs/'+datas[0].select('td.subject')[0].select_one('a')['href']
  result.append([title,link_url])

result
```
