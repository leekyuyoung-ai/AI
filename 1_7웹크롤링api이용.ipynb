```python
import json
import pandas as pd

url = 'http://apis.data.go.kr/1360000/MidFcstInfoService/getMidFcst'
service_key = '여러분들의 일반 인증키 key'

# 요청 파라메터
params ={
    'serviceKey':service_key,
    'numOfRows' : 10,
    'pageNo' : 1,
    'stnId' : 109,  # 서울 인천 경기
    'tmFc' : 202411270600,
    'dataType' : 'JSON'
}
# get 요청
response = requests.get(url, params=params)
print(f'응답상태 코드 : {response.status_code}')
# print(response.text)  # json 은 dict 형식이지만 문자열
# 보기 좋게 출력
data = response.json()  # json 파싱
# print(json.dumps(data, indent=2, ensure_ascii=False))  # 보기좋게 들여쓰기 4만큼 모양으로 출력
print(data['response']['body']['items']['item'][0]['wfSv'])
```

카카오 API
  - 주소를 위 경도로 변환해 주는

```python
'''
curl -v -G GET "https://dapi.kakao.com/v2/local/search/address.json" \
  -H "Authorization: KakaoAK ${REST_API_KEY}" \
  --data-urlencode "query=전북 삼성동 100"
'''
def findXY(address):
  url = 'https://dapi.kakao.com/v2/local/search/address.json'
  REST_API_KEY = '발급받은 카카오 api키 사용'
  headers = {
      "Authorization": f"KakaoAK {REST_API_KEY}"
  }
  params = {
      "query" : address
  }
  response = requests.get(url, headers=headers, params=params)
  # print(response.status_code)
  data = response.json()
  # print(json.dumps(data, indent=2, ensure_ascii=False))  # 보기좋게 출력
  #  위 경도를 찾아서 지도에 표시
  x = data['documents'][0]['address']['y']  # 위도
  y = data['documents'][0]['address']['x']  # 경도
  return x, y
```

```python
findXY("서울특별시 강남구 강남대로94길 13 삼경빌딩")
```

```python
import folium
x, y = findXY("서울특별시 강남구 강남대로94길 13 삼경빌딩")
map = folium.Map(location = [x,y])  # 지도 생성 (x, y) 중심으로 하는 지도
folium.Marker(
    location=[x,y],
    popup='삼경빌딩',
    icon=folium.Icon(color='red', icon='info-sign')
).add_to(map)   # 마커를 map에 추가
map  # 지도 실행
```

```python
import pandas as pd
df = pd.read_csv('/content/서울시 대형학원 정보.csv', encoding='cp949')
df = df.iloc[:,[1,3]]  # 이름하고 도로명 주소
df.dropna(inplace=True)
df
# 도로명 주소에서 위 경도를 추출해서 지도에 전부 마킹
# 지도를 만들고 우리학원을 기준으로
map = folium.Map(location = ['37.4995844298918', '127.028419603321'])  # 지도 생성 (x, y) 중심으로 하는 지도

df.reset_index(drop=True, inplace=True)


for index in range(len(df)):
  # 각 주소의 위경도를 구해서
  try:
    x, y = findXY(df.iloc[index, 1])
    # 우리 지도에 마킹
    folium.Marker(
      location=[x,y],
      popup=df.iloc[index, 0],
      icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(map)   # 마커를 map에 추가
  except:
    pass
map
```
